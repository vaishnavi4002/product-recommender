```python
import pandas as pd
import numpy as np
products= pd.read_csv('amazon.csv')
```

```python
products.head()
```

```python
product_info = products[['product_name', 'rating', 'rating_count','actual_price','img_link']]
product_info['rating_count'] = pd.to_numeric(product_info['rating_count'], errors='coerce')

product_info = product_info[product_info['rating_count'] > 500]
product_info
```

```python
# Convert the 'rating' column to numeric, coercing errors to NaN
product_info['rating'] = pd.to_numeric(product_info['rating'], errors='coerce')

# Define the aggregation function for each column
aggregation_functions = {
    'rating': 'mean',
    'actual_price': 'first', 
    'img_link': 'first' 
}

# Group by 'product_name' and apply the aggregation functions
product_info = product_info.groupby('product_name').agg(aggregation_functions).reset_index()
```

```python
popular_products=product_info.head(50).sort_values('rating',ascending=False)
popular_products.head(3)['img_link']='https://m.media-amazon.com/images/I/71U7owVHbaL._SX466_.jpg'
popular_products.head(4)['img_link']='https://m.media-amazon.com/images/I/61AK6enlSIL._SX679_.jpg'
```

```python
import pickle
pickle.dump(popular_products,open('popular.pkl','wb'))
```
