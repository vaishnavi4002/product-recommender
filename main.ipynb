```python
import pandas as pd
import numpy as np
products= pd.read_csv('amazon.csv')

products.head(5)
```

```python
products['user_ids'] = products['user_id'].str.split(',')
```

```python
products_expanded = products.explode('user_ids')
```

```python
products_expanded.drop(columns=['user_id'], inplace=True)
```

```python
products_expanded.reset_index(drop=True, inplace=True)
products_expanded.head()
```

```python
products_expanded.shape
```

```python
user_rating_counts = products_expanded['user_ids'].value_counts()
user_rating_counts
```

```python
experienced_user_ids = user_rating_counts[user_rating_counts >=4].index
experienced_user_ids 
products_filtered = products_expanded[products_expanded['user_ids'].isin(experienced_user_ids )]
products_filtered
```

```python
unique_products = products_filtered.groupby('product_name').agg({'user_ids': lambda x: list(set(x)), 'rating': 'first'}).reset_index()
unique_products
```

```python
# Convert lists in the 'user_ids' column to tuples
unique_products['user_ids'] = unique_products['user_ids'].apply(tuple)

# Drop duplicate rows based on the 'product_name' column
unique_products = unique_products.drop_duplicates(subset=['product_name'], keep='first', ignore_index=True)

# Display the resulting DataFrame
unique_products



```

```python
# Explode the 'user_ids' column to create separate rows for each user ID
exploded_products = unique_products.explode('user_ids')

# Pivot the table
pt = exploded_products.pivot_table(index='product_name', columns='user_ids', values='rating', aggfunc='first')

# Fill NaN values with 0
pt.fillna(0, inplace=True)

pt = pt.iloc[2:]
# Display the pivot table
pt
```

```python
pt

```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_score = cosine_similarity(pt)
```

```python
def recommend(partial_name):
    matching_products = [product for product in pt.index if partial_name.lower() in product.lower()]
    
    data = []

    if matching_products:
        product_name = matching_products[0]  # Selecting the first matched product
        index = pt.index.get_loc(product_name)
        similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
        print("Similar products for '{}':".format(product_name))
        
        for i, (product_index, score) in enumerate(similar_items, 1):
            item = []
            temp_df = products[products['product_name'] == pt.index[product_index]]
            item.append(temp_df.drop_duplicates('product_name')['product_name'].values[0])
            item.append(temp_df.drop_duplicates('product_name')['rating'].values[0])
            item.append(temp_df.drop_duplicates('product_name')['actual_price'].values[0])
            item.append(temp_df.drop_duplicates('product_name')['img_link'].values[0])
            data.append(item)

    else:
        print("No products found matching the specified partial name.")

    return data
```

```python
recommend('cable')
```
